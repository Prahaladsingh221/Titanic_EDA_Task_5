```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Sex'].value_counts()
df['Pclass'].value_counts()
df['Embarked'].value_counts()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)
```

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title('Gender Count')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.pairplot(df[['Survived','Pclass','Age','Fare']], hue='Survived')
```

## Summary of Insights:

- Most passengers were between 2040 years old.
- More male passengers than female.
- Fewer people survived than died.
- Women and 1st class passengers had a better survival rate.
- Fare had a positive correlation with survival.
